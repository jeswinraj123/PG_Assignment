```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("application_data.csv")
data
```

```python
data.columns
```

```python
late_installment = data[data['TARGET'] == 1]
late_installment
```

```python
print(len(data),len(late_installment))
```

```python
late_installment['AMT_INCOME_TOTAL'].value_counts()
```

```python
late_installment.hist(column='AMT_CREDIT')
```

```python
fig  = plt.figure(figsize=(20,10),facecolor="lightgrey")
ax = plt.axes()
ax.set_facecolor("silver")
plt.title("Histogram of AMOUNT CREDITED",fontdict={"size":20})
plt.xlabel("Count",fontdict={"size":20})
plt.ylabel("AMOUNT CREDITED in million Dollars",fontdict={"size":20})
plt.tick_params('both',size=10,labelsize=15)
plt.hist(late_installment['AMT_CREDIT']/1000000,bins=10,color="lightblue",histtype="barstacked",orientation="horizontal")
```

```python
late_installment = late_installment[late_installment['AMT_INCOME_TOTAL']<12000000]
```

```python
income = late_installment.sort_values(by='AMT_INCOME_TOTAL')
fig  = plt.figure(figsize=(20,10),facecolor="lightgrey")
ax = plt.axes()
ax.set_facecolor("lightcyan")
plt.title("PLOT Between Income and Loan Amount Credited",fontdict={"color":"darkgreen","size":15})
plt.xlabel("ANNUAL INCOME",fontdict={"color":"darkgreen","size":15})
plt.ylabel("LOAN AMOUNT CREDITED",fontdict={"color":"darkgreen","size":15})
plt.tick_params('both',size=10,labelsize=15)
plt.plot(income['AMT_INCOME_TOTAL']/1000000,income['AMT_CREDIT']/1000000,color='plum')
```

```python
fig  = plt.figure(figsize=(20,10),facecolor="lightgrey")
ax = plt.axes()
ax.set_facecolor("lightcyan")
plt.title("BOX PLOT OF LOAN AMOUNT",fontdict={"color":"darkgreen","size":15})
plt.xlabel("LOAN AMOUNT",fontdict={"color":"darkgreen","size":15})
plt.tick_params('both',size=10,labelsize=15)
plt.boxplot(income['AMT_CREDIT'],patch_artist=True,boxprops={"facecolor":"lawngreen","color":"yellowgreen"},capwidths=0.1)
```

```python
owns_car_or_house = (late_installment['FLAG_OWN_REALTY']+late_installment['FLAG_OWN_CAR']).astype("category").cat.codes
late_installment.plot(kind="scatter",x='AMT_INCOME_TOTAL',y='AMT_GOODS_PRICE',cmap=plt.get_cmap("spring"),c=owns_car_or_house)
```

```python
fig  = plt.figure(figsize=(20,10),facecolor="lightgrey")
ax = plt.axes()
ax.set_facecolor("lightcyan")
plt.title("Bar Chart of Late installement Contract Type",fontdict={"color":"darkgreen","size":15})
colors = sns.color_palette('pastel')[0:5]
value_count = late_installment['NAME_CONTRACT_TYPE'].value_counts()
plt.pie(x=value_count.values,labels=list(value_count.index), colors = colors, autopct='%.0f%%')
```
